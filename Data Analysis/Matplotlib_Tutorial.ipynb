# **Matplotlib Tutorial**

### Importing Library

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

### Basic Graph

```python
## Line 1
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]

## Resize the graph
plt.figure(figsize = (5, 3), dpi = 120)

## Keyword Argument Notation
# plt.plot(x, y, label = '2x', color = 'red', linewidth = 3, linestyle = '--', marker = '.', markersize = 15, markeredgecolor = 'blue')

## Shorthand notation
## fmt = '[color][marker][line]'
plt.plot(x, y, 'r.--', label = '2x', markersize = 15, markeredgecolor = 'blue', linewidth = 3)

## Line 2

## Select interval we want to plot points at
x2 = np.arange(0, 4.5, 0.5)

## Plot part of the graph as line
plt.plot(x2[:5], x2[:5]**2, 'b',label = 'x^2')

## Plot remaider of graph as dot
plt.plot(x2[4:], x2[4:]**2, 'b--')

## Add a title
plt.title('My First Graph!', fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})

## X and Y Labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

## Scaling the grapgh
plt.xticks([0, 1, 2, 3, 4])
plt.yticks([0, 2, 4, 6, 8, 10])

## add a legend
plt.legend()

## Save Figure
plt.savefig('mygraph.png', dpi = 300)

## Show Plot
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1, 4, 2]

bars = plt.bar(labels, values)

patterns = ['/', 'O', '*']
for bar in bars:
    bar.set_hatch(patterns.pop(0))

# bars[0].set_hatch('/')
# bars[1].set_hatch('O')
# bars[2].set_hatch('*')

plt.figure(figsize = (6, 4))

plt.show()
```
