### Loading data into Pandas

```python
df = pd.read_csv('pokemon_data.csv')
print(df.head(10))
```

### Reading data in Pandas

```python
## Read Headers
# df.columns

## Read each Column
# df[['Name', 'Type 1', 'HP']][0:10]

## Read each Row
# print(df.iloc[0:4])

# for index, row in df.iterrows():
#     print(index, row['Name'])

df.loc[df['Type 1'] == "Fire"]

## Read a specific Location(R,C)
# print(df.iloc[2, 1])
```

### Sorting/Describing Data

```python
df.describe()
```

```python
df.sort_values(['Type 1', 'HP'], ascending=[1, 0])
```

### Making changes to the data

```python
# df['Total'] = df['HP'] + df['Attack'] + df['Defense'] + df['Sp. Atk'] + df['Sp. Def'] + df['Speed']
# df = df.drop(columns=['Total'])

df['Total'] = df.iloc[:, 4:10].sum(axis=1)
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]

df.head(5)
```

### Saving our Data

```python
df.to_csv('modified.csv', index=False)
```

### Filtering Data

```python
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]
# new_df.to_csv('filtered.csv')

new_df.reset_index(drop = True, inplace = True)
new_df
```

```python
df.loc[~df['Name'].str.contains('Mega')]
```

```python
import re
df.loc[df['Type 1'].str.contains('Fire|grass', flags=re.IGNORECASE, regex=True)]
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.IGNORECASE, regex=True)]
```

### Conditional Changes

```python
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flammer'
df
```

```python
df.loc[df['Type 1'] == 'Flammer', 'Type 1'] = 'Fire'
df
```

### Aggregate Statistics(Groupby)

```python
df = pd.read_csv('modified.csv')
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

```python
df['Count'] = 1
count_df = df.groupby(['Type 1', 'Type 2']).count()['Count']
count_df.to_csv('Count.csv')
count_df
```

### Working with large amounts of data

```python
for df in pd.read_csv('modified.csv', chunksize=5):
    print("**********")
    print(df)
```

```python
new_df1 = pd.DataFrame(columns = df.columns)

for df in pd.read_csv('modified.csv', chunksize=5):
    results = df.groupby(['Type 1']).count()

    new_df1 = pd.concat([new_df1, results])
```
